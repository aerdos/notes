```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def gaussian(x, mu, sig):
    return (1.0 / (sig * np.sqrt(2.0 * np.pi))) * np.exp(-(1.0/2.0) * (((x - mu)/sig) ** 2.0))
```

```python
xx = np.linspace(0, 15, 201)
T = xx[1] - xx[0]
```

```python
P_est_1 = gaussian(xx, 3, np.sqrt(0.25))
P_est_2 = gaussian(xx, 9, np.sqrt(0.6))
P_meas_1 = gaussian(xx, 10, np.sqrt(0.4))
P_est_3 = P_est_2 * P_meas_1
P_est_3_norm = P_est_3 / (np.sum(P_est_3 * T))
```

```python
ii_max = np.argmax(P_est_3_norm)
yy_max = P_est_3_norm[ii_max]
```

```python
xx[ii_max]
```

```python
f = plt.figure(figsize=(12, 5))
ax1 = f.add_subplot(111)


ax1.plot(xx, P_est_1, color='xkcd:blue', label='Model Est k-1')
ax1.plot(xx, P_est_2, color='xkcd:blue', label='Model Est k')
ax1.plot(xx, P_meas_1, color='xkcd:orange', label='GPS Meas k')
# ax1.plot(xx, P_est_3, color='xkcd:green', label='Pos Est k')
ax1.plot(xx, P_est_3_norm, color='xkcd:green', label='Pos Est k')
ax1.scatter(xx[ii_max], yy_max)

ax1.legend()
ax1.set_xlabel('Position')
ax1.set_ylabel('Probability of Measurement')
ax1.set_xlim([0, 15])
ax1.set_ylim([0, 1.5])
```
